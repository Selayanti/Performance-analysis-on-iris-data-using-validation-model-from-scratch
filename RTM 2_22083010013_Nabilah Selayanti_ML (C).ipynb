```python
import pandas as pd
import numpy as np
data = pd.read_excel("data iris.xlsx")
data
```

```python
# Membuat dictionary untuk mapping label kelas ke angka
label_mapping = {'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 3}

# Mengganti nilai kelas dengan angka menggunakan mapping
data['CLASS'] = data['CLASS'].map(label_mapping)

# Menampilkan DataFrame setelah perubahan
data
```

```python
# Inisialisasi list untuk menyimpan data latih dan data uji
train_data = []
test_data = []

# Loop untuk LOO
for i in range(len(data)):  # Iterasi hingga iterasi ke-149
    # Keluarkan satu sampel sebagai data uji
    test_sample = data.iloc[i:i+1]
    
    # Buat data latih tanpa sampel yang dikeluarkan
    train_samples = data.drop(index=i)
    
    # Tambahkan data ke list
    train_data.append(train_samples)
    test_data.append(test_sample)

# Konversi list data latih dan data uji menjadi DataFrame
train_dfs = [pd.DataFrame(train) for train in train_data]
test_dfs = [pd.DataFrame(test) for test in test_data]

# Output data latih dan data uji dalam bentuk array atau matriks
for i in range(len(train_data)):
    print(f"Iterasi ke-{i+1}:")
    print("Data Train:")
    print(train_data[i])
    print("\nData Test:")
    print(test_data[i])
    print()
```

```python
# Inisialisasi list untuk menyimpan hasil
euclidean_distances_all_iterations = []

# Loop untuk LOO
for i in range(len(data)):
    # Ambil data uji
    test_sample = data.iloc[i]
    # Ambil data latih
    train_samples = data.drop(index=i)
    
    # Inisialisasi list untuk menyimpan Euclidean distance untuk sampel ini
    sample_distances = []
    
    # Loop untuk menghitung Euclidean distance antara data uji dan setiap sampel data latih
    for _, train_sample in train_samples.iterrows():
        # Hitung Euclidean distance
        euclidean_distance = ((test_sample['X1'] - train_sample['X1']) ** 2 + 
                              (test_sample['X2'] - train_sample['X2']) ** 2 + 
                              (test_sample['X3'] - train_sample['X3']) ** 2 + 
                              (test_sample['X4'] - train_sample['X4']) ** 2) ** 0.5
        sample_distances.append(euclidean_distance)
    
    # Tambahkan hasil Euclidean distance untuk sampel ini ke list utama
    euclidean_distances_all_iterations.append(sample_distances)

# Membuat DataFrame dari hasil
euclidean_matrix = pd.DataFrame(euclidean_distances_all_iterations)

# Menambahkan kolom untuk iterasi
euclidean_matrix.columns = [f"Tetangga {i+1}" for i in range(len(result_df.columns))]
euclidean_matrix.index = [f"Iterasi ke-{i+1}" for i in range(len(result_df))]

# Menampilkan DataFrame
euclidean_matrix
```

```python
# Mencari 5 tetangga terdekat untuk setiap iterasi
nearest_neighbors_all_iterations = []

for distances_iter in euclidean_distances_all_iterations:
    # Mengurutkan hasil euclidean distance
    sorted_distances = sorted(distances_iter)
    # Memilih 5 nilai terkecil (tetangga terdekat)
    nearest_neighbors = sorted_distances[:5]
    nearest_neighbors_all_iterations.append(nearest_neighbors)

# Membuat DataFrame dari hasil
df_nearest_neighbors = pd.DataFrame(nearest_neighbors_all_iterations, columns=["Tetangga 1", "Tetangga 2", "Tetangga 3", "Tetangga 4", "Tetangga 5"])

# Menambahkan kolom iterasi
df_nearest_neighbors.insert(0, "Iterasi", range(1, len(df_nearest_neighbors) + 1))
df_nearest_neighbors
```

```python
# Mencari label 'class' dari masing-masing nilai pada 5 tetangga terkecil
predicted_labels_all_iterations = []

for distances_iter in euclidean_distances_all_iterations:
    # Mengambil 5 tetangga terdekat
    nearest_indices = np.argsort(distances_iter)[:5]
    nearest_labels = [train_samples.iloc[idx]['CLASS'] for idx in nearest_indices]
    
    # Mencari kelas yang paling sering muncul di antara tetangga terdekat
    predicted_label = max(set(nearest_labels), key=nearest_labels.count)
    predicted_labels_all_iterations.append(predicted_label)

# Mengubah tipe data kolom 'Predicted Class' menjadi integer
df_nearest_neighbors['Predicted Class'] = df_nearest_neighbors['Predicted Class'].astype(int)

# Cetak hasil prediksi setelah mengubah format nilai
df_nearest_neighbors
```

```python
# Inisialisasi list untuk menyimpan nilai miss match dan iterasi ke berapanya
miss_match_info = []

# Loop melalui setiap baris hasil prediksi dan data pelatihan
for i, (pred_label, train_label) in enumerate(zip(df_nearest_neighbors['Predicted Class'], train_samples['CLASS']), start=1):
    # Bandingkan nilai prediksi dengan nilai aktual
    if pred_label != train_label:
        # Jika terjadi miss match, tambahkan informasi iterasi dan indeks baris ke dalam list miss_match_info
        miss_match_info.append((i, pred_label, train_label))

# Menampilkan informasi miss match
print("Informasi miss match:")
for info in miss_match_info:
    print(f"Iterasi ke-{info[0]}: Prediksi={info[1]}, Aktual={info[2]}")

# Menampilkan jumlah miss match
print("Jumlah miss match:", len(miss_match_info))
```

```python
# Menghitung jumlah data tes
jumlah_data_test = len(test_data)

# Menghitung jumlah miss match
jumlah_miss_match = len(miss_match_info)

# Menghitung akurasi
akurasi = ((jumlah_data_test - jumlah_miss_match) / jumlah_data_test) * 100

# Menampilkan akurasi
print("Akurasi:", akurasi, "%")
```

```python
# Menghitung total rata-rata error
total_rata_rata_error = len(miss_match_info) / 150

# Menampilkan total rata-rata error
print("Total Rata-rata Error:", total_rata_rata_error, "%")
```
